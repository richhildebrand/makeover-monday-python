```python
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
def draw_colorbar(header_ax, fig, data, cmap):
    colorbar_ax = fig.add_axes([0.775, 0.77, 0.125, 0.1])
    colorbar_im = colorbar_ax.imshow(data, cmap=cmap, aspect=0.125)
    colorbar = fig.colorbar(colorbar_im, cax=colorbar_ax, orientation='horizontal')
```

## Create two subpots

1. We want a header area to place our colorbar in
2. We want a seperate area to place our heatmap in

*This will prevent our colorbar from affecting the size ratio of our heatmap.*

```python
fig, (header_ax, heatmap_ax) = plt.subplots(nrows=2, figsize=(15, 10), gridspec_kw={"height_ratios":[0.1, 1]})

cmap = 'coolwarm'
data = np.random.rand(10, 15)
sns.heatmap(data, ax=heatmap_ax, cmap=cmap, cbar=False)
header_ax.axis('off') #remove grid from header subplot

draw_colorbar(header_ax, fig, data, cmap)
```
